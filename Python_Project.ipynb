Bank Chatbot 



## Objective: 

Goal: Build a chat bot that can greet the user and allow users to access their bank account and do the following: 
1.	Inquire about the balance  
2.	Make a deposit 
3.	Make a withdrawal 
4.	Close the account 
5.	Give the user the current weather and time 

**Further Instructions:** The chat bot should be constructed using classes and objects. The weather should be retrieved using an api call to a weather site with a public key. 


```python
# A bankaccount chat bot

import requests
import datetime
import pytz

#Creating a class of bank account with methods DEPOSIT,WITHDRAW,BALANCE CHECK
class BankAccount:
  def __init__(self,name,balance):
    self.name = name
    self.balance = balance
  

  def deposit(self,amount): # method to deposit
    self.balance += amount

  def withdraw(self,amount):# method to withdraw 
    if amount > self.balance:
      return "Insufficient balance"
    else:
      self.balance -= amount
  
  def check_balance(self): # method to check balance
    return self.balance


# Creating a chatbot class with methods to greet,get_time and get_weather
class ChatBot:
  def __init__(self,API_KEY,timezone='Asia/kolkata'): # we can change the timezone accordingly like 'Asia/kolkata' or 'Europe/London'
    self._API_KEY = API_KEY
    self.greeting = 'Hello, Welcome to chatbot.How can I help you today?'
    self._location= ""
    self._timezone = timezone
    
  #to greet the user
  def greet(self):
    return self.greeting

  #gets the current time using the datetime module. here the timezone is a private attribute
  def set_timezone(self, timezone):
        self._timezone = timezone

  # to get the time of the specific timezone using pytz module
  def get_time(self):
        tz = pytz.timezone(self._timezone)
        current_time = datetime.datetime.now(tz).strftime("%H:%M:%S %p")
        return f"The current time in your location is :{current_time}"
    
  #gets the current weather of a location using the provided api key and location
  def get_weather(self):
    url= f"http://api.openweathermap.org/data/2.5/weather?q={self._location}&appid={self._API_KEY}"
    response = requests.get(url)
    weather = response.json()
    temperature = round(weather['main']['temp'] -273.15) # to get the temperature in celsius
    humidity = weather['main']['humidity']
    return f"The current weather in {location} is {weather['weather'][0]['description']}, with a temperature of {temperature}°C and humidity of {humidity}%"

  #creates a private attribute for location that is changeble
  def set_location(self,location):
    self._location=location


#creating instance for chatbot 
API_KEY ='6fe345b29fac6ac5e4ea4ed9c480331c' #  A public key to connect with the weathermap
bot = ChatBot(API_KEY) 
print(bot.greet())

name = input(' Please enter your name:')
balance=int(input(f' Mr./Ms./Mrs.{name}, please enter your initial balance :INR '))


#creating an instance for  class BankAccount
account = BankAccount(name,balance)

# Loop to continue getting user input till the user types 'exit'
while True:
  action = input(f'Mr./Ms./Mrs.{name},What would you like today?(inquire/deposit/withdraw/weather/time/exit:')

  #to check balance
  if action.lower() == 'inquire':
    print('Your current balance is : INR ',account.check_balance())

  #to deposit
  elif action.lower() == 'deposit':
    amount = int(input('Please enter the amount you would like to deposit:INR '))
    account.deposit(amount)
    print('Your updated balance is :INR',account.check_balance())

  #to withdraw
  elif action.lower() == 'withdraw':
    amount = int(input('Please enter the amount you would like to withdraw:INR '))
    result = account.withdraw(amount)
    if result=='insufficient balance':
      print(result)
    else:
      print('Your updated balance is:INR ',account.check_balance())
  
  #to get weather info
  elif action.lower() == 'weather':
    location = input('Enter the location name :')
    bot.set_location(location)
    print(bot.get_weather())

  #to get time
  elif action.lower() == 'time':
    print(bot.get_time())

  #to exit
  elif action.lower() == 'exit':
    print('Goodbye,see you next time!!')
    break
  
  else:
    print('Invalid option selected,please try again')





    
  
```
